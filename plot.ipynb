```python
import numpy as np
import matplotlib.pyplot as plt

def get_data(file_name):
    idx = []
    fitness = []
    with open(file_name, 'r') as f:
        lines = f.read().splitlines()
    lines = lines[:1000]
    for line in lines:
        # print('--'+line+'--')
        line = line.split(' ', 1)
        idx.append(int(line[0]))
        fitness.append(float(line[1]))
    return idx, fitness
```

```python
file = ['fitness/gset1.txt', 'fitness/gset1-re.txt', 'fitness/gset1-mu.txt']
color = ['red', 'blue', 'green', 'purple']
label = ['task1', 'task1-re', 'task1-mu']
for i in range(3):
    x, y = get_data(file[i])
    plt.plot(x, y, label=label[i], c=color[i], linewidth=1)
plt.legend(loc="lower right")
plt.title("Fitness change in iteration (gset-id = 1)")
plt.xlabel("Iteration time")
plt.ylabel("Fitness")
plt.show()
```
